# Apriori Algorithm

figure out the variuos items present in transaction data.
and which are the 2 items which can be sold together

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('store_data.csv', header=None)
df
```

```python
df.info()
```

```python
# At max total 20 items in a particular transaction
# convert above data frame to list because this algorithm works on list 
records = []

for i in range(0,7501):
    records.append([str(df.values[i,j]) for j in range(0,20)])

print(records[:4])
```

```python
for row  in df.iterrows():
    print(list(row))
    break
```

```python
# Use the Apriori Algorithm
#!pip install apyori
```

```python
from apyori import apriori
```

```python
association_rules = apriori(records, min_lift=3, max_length=2, min_confidence=0.2, min_support=0.0045) # min_lift=3 atleast 3 times sales should increase with 
#max_length=2 (2 products that can bundled together)
# support is very small because a particular item might be just bought in 2 transaction (very less)
```

```python
#list(association_rules)
```

```python
# Interpret
# RelationRecord(items=frozenset({'pasta', 'shrimp'}), support=0.005065991201173177, ordered_statistics=[OrderedStatistic(items_base=frozenset({'pasta'}), items_add=frozenset({'shrimp'}), confidence=0.3220338983050847, lift=4.506672147735896)])]
# here lift=4.506672147735896 indicate that the combo of pasta and shrimp would increase the sales by 4.5 times if sold together
```

```python
for rule in association_rules:
    print(rule.items)
    print(f" Lift : {rule.lift}")
    print(f" confidence : {rule.confidence}")
    print(f" support : {rule.support}")
    break
```
