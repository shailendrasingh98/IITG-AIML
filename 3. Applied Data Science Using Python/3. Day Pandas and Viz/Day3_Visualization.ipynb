```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Simple plot in x y plane
x = range(1,6,1)
y =range(2,12, 2)
plt.plot(x,y)
plt.show()
```

```python
# Simple plot in x y plane
x = range(1,6,1)
y = range(2,12, 2)

# Adjust the size of plot
plt.figure(figsize=(20,5))


# add color and label for plot
plt.scatter(x,y, color="Green", label="Scatter plot")
# to show the legend
plt.legend() # by default label is taken as legend but we can print different legend like .legend('d')
# to show x and y label
plt.xlabel("x-data")
plt.ylabel("y-data")
# to show grid 
plt.grid()
# print the plot
plt.show()
```

```python
# this plot is used for finding the outlier.
# Here dots (-400, 1000) represent the otlier
# uni-variate
var = [100,200,300,500,-400,1000,322,112,124,400]
plt.boxplot(var) # this plot is used for finding the outlier
plt.ylabel('data')
plt.show()
```

```python
# Bar chart to show the categorical data
categories = ['A', 'B', 'C']
values = [10,20,4]

plt.bar(categories,values,color='orange', edgecolor='green', label= 'Random_data')

# To show Title
plt.title('Bar char: 3 Categories')
plt.legend()
plt.xlabel('categories')
plt.ylabel('Values')
plt.show()
```

```python
# data is list of 3 numpy arrays
data = [np.random.normal(0,1,100),np.random.normal(0,1.5,100),np.random.normal(0,2,200)]

#
plt.boxplot(data, labels=['A', 'B', 'C'])
# To show Title
plt.title('Boxplot Group comparision')
plt.legend()
plt.xlabel('Groups')
plt.ylabel('Values')
plt.show()
```

```python
# this same as above but 
plt.boxplot(data, labels=['A', 'B', 'C'],patch_artist=True, notch=True)
# To show Title
plt.title('Boxplot Group comparision')
plt.legend()
plt.xlabel('Groups')
plt.ylabel('Values')
plt.show()
```

```python
# random data for radar chart
categories = ['A', 'B', 'C', 'D', 'E']
data = np.random.randint(0,10,len(categories))

# create a radar chart
angles = np.linspace(0,2*np.pi, len(categories), endpoint =False)
data = np.concatenate((data, [data[0]]))
angles = np.concatenate((angles, [angles[0]]))

# polar plot
plt.polar(angles, data)

# add lables
plt.thetagrids(np.degrees(angles[:-1]),labels=categories)
plt.title('Polar  comparision')

plt.show()
```
