```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import bernoulli, gaussian_kde
# Bernoulli distribution parameters
p = 0.6  # Probability of success
# Generating Bernoulli distribution data
data_bernoulli = bernoulli.rvs(p, size=1000)
# Plot histogram
plt.hist(data_bernoulli, bins=np.arange(-0.5, 2, 1), density=True, alpha=0.5, color='pink', edgecolor='black', linewidth=2, align='mid')
# Calculate the KDE
kde = gaussian_kde(data_bernoulli)
kde_xs = np.linspace(-1, 2, 300)
kde_ys = kde.pdf(kde_xs)
# Plot KDE
plt.plot(kde_xs, kde_ys, color='red')
# Add labels
plt.xlabel('Bernoulli')
plt.ylabel('Frequency')
# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 
from scipy.stats import bernoulli 
# Bernoulli distribution parameters
p = 0.6 # Probability of success 
# Generating Bernoulli distribution data 
data_bernoulli = bernoulli.rvs(p, size=1000) # Unique values (0 and 1) and their probabilities
#probability mass function 
x = [0, 1] 
pmf_values = bernoulli.pmf(x, p) # PMF for Bernoulli 
# Plot histogram (empirical data) 
plt.hist(data_bernoulli, bins=np.arange(-0.5, 2, 1), density=True, alpha=0.5, color='pink', edgecolor='black', linewidth=2, align='mid', label="Empirical Data") 
# Plot PMF (theoretical probabilities)
plt.scatter(x, pmf_values, color='red', marker='o', s=100, label="PMF (Theoretical)") 
# Labels and legend 
plt.xlabel('Bernoulli Outcome') 
plt.ylabel('Probability') 
plt.xticks([0, 1]) 
plt.legend() 
plt.title("Bernoulli Distribution (p=0.6)") 
# Display the plot 
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

# Parameters for the binomial distribution
n, p = 20, 0.8  # number of trials, probability of success per trial

# Generate binomially distributed data
data_binom = binom.rvs(n, p, size=10000)

# Plot histogram
plt.hist(data_binom, bins=range(n+2), density=True, alpha=0.5, color='green', label="Histogram")

# Calculate and plot PMF
x = np.arange(n+1)
pmf_values = binom.pmf(x, n, p)
plt.scatter(x, pmf_values, color='red', label="Binomial PMF", zorder=3)

# Add labels
plt.xlabel('Binomial')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

### Poisson

The Poisson distribution is a probability distribution used to model the number of events that occur within a fixed interval of time or space, given that the events happen independently and the average rate of occurrence is constant.

**Key Characteristics of the Poisson Distribution:**
*Discrete Distribution*: It models the count of events (0, 1, 2, 3, etc.) that occur in a fixed interval.
Constant Rate: The events occur at a constant average rate, denoted by ùúÜ.

- Œª (lambda), which represents the expected number of occurrences in the given interval.
- Independence: The events happen independently of each other. The occurrence of one event does not affect the probability of another event occurring.
- Rare Events: The Poisson distribution is often used for modeling rare or infrequent events over a continuous interval.

Example Use Cases:
- Number of phone calls received by a call center in an hour.
- Number of emails you receive in a day.
- Number of accidents at a busy intersection over a week.
- Number of defective items found in a manufacturing process over a certain period (if the defects are rare and independent).

**Poisson vs. Binomial:**
*Binomial:* Used when there are a fixed number of trials, each with two possible outcomes (success/failure), and the probability of success is the same for each trial.

*Poisson:* Used when there‚Äôs no fixed number of trials, but rather a continuous time or space interval, with a focus on the number of events occurring in that interval at a constant rate.

- The Poisson distribution can be thought of as a limiting case of the binomial distribution when:

- The number of trials becomes very large.

- The probability of success is very small.
- The product of the number of trials and the probability of success remains constant.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import poisson

# Parameter for Poisson distribution
lambda_ = 10  # Mean number of events

# Generate Poisson-distributed data
data_poisson = poisson.rvs(lambda_, size=10000)

# Plot histogram
plt.hist(data_poisson, bins=range(min(data_poisson), max(data_poisson) + 2),
         density=True, alpha=0.5, color='blue', label="Histogram")

# Calculate and plot PMF
x = np.arange(min(data_poisson), max(data_poisson) + 1)
pmf_values = poisson.pmf(x, lambda_)
plt.scatter(x, pmf_values, color='red', label="Poisson PMF", zorder=3)

# Add labels and legend
plt.xlabel('Poisson')
plt.ylabel('Probability')
plt.legend()
plt.show()
```
