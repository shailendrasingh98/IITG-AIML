# Preprocessing

StandardScaler is a preprocessing tool in Python's sklearn.preprocessing module that standardizes features by removing the mean and scaling to unit variance. It is commonly used in machine learning pipelines to ensure that features have a standard normal distribution, which helps many algorithms perform optimally.

How StandardScaler Works
The transformation is performed using the formula:

z = x-μ/σ
 
Where:

- x: Original feature value
- μ: Mean of the feature values
- σ: Standard deviation of the feature values

This results in a dataset where each feature has:

- A mean (μ) of 0

- A standard deviation (σ) of 1

**Key Parameters**
- with_mean: If True (default), centers the data by subtracting the mean.

- with_std: If True (default), scales the data to unit variance.

**When to Use StandardScaler**
Algorithms Sensitive to Feature Scaling:
Algorithms like Support Vector Machines (SVM), Principal Component Analysis (PCA), and Gradient Descent-based models require standardized data for optimal performance.

Features with Different Scales:
When features have varying units or ranges, standardization ensures fair treatment during modeling.

```python
from sklearn.preprocessing import StandardScaler
```

**MinMaxScaler:**
Scaling technique used in machine learning to normalize data within a specified range, typically between 0 and 1.

This scaling method is particularly useful for algorithms that rely on distance metrics, such as K-Nearest Neighbors (KNN) and K-Means clustering.
- 1. Standardize Data:

-----> Xstd = Xmax −Xmin / X−Xmin
 
- 2. Scale to Desired Range: 

-----> Xscaled =Xstd ×(max−min)+min

Here, min and max are the desired range limits, defaulting to (0, 1).


```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max = MinMaxScaler()
df['table'] = min_max.fit_transform(df[['table']])
df
```

```python
df = pd.read_csv('diamonds.csv', index_col='Unnamed: 0')

numeric_columns = []
not_numeric_columns = []
overall_columns = df.columns
print(f" All columns: {overall_columns}")
# Find numeric and non numeric column
for col in overall_columns:
    if df[col].dtype =='O':
        not_numeric_columns.append(col)
    else:
        numeric_columns.append(col)

print(f" Numeric columns: {numeric_columns}")
print(f" Non numeric columns: {not_numeric_columns}")
```

```python
# Separate numeric and non-numeric columns using select_dtypes
numeric_columns = df.select_dtypes(include=['number']).columns.tolist()
non_numeric_columns = df.select_dtypes(exclude=['number']).columns.tolist()
print(f" Numeric columns: {numeric_columns}")
print(f" Non numeric columns: {not_numeric_columns}")
```

```python
new_df1 = df[numeric_columns]
new_df2 = df[non_numeric_columns]
print(new_df1)
print('-'*50)
print(new_df2)
```

```python
pd.concat([new_df1, new_df2]) # Here axis = 0 so data is concated column wise
```

```python
final_df = pd.concat([new_df1, new_df2],axis =1) # axis =1 is for row wise
final_df
```

```python
# define bin the edges and labels
segments= [0, 0.3, 0.5, df['carat'].max()] # Bin Edges
category = ['Small', 'Medium', 'Large']
final_df['carat_category'] = pd.cut(final_df['carat'], bins=segments, labels = category)
final_df[['carat','carat_category']].head(100)

```

## Outliers

```python
df = pd.read_csv("diamonds.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
df['carat'].describe()
```

```python
plt.boxplot(df['carat'])
plt.ylabel('carat')
plt.show()
```

```python

Q1_carat = df['carat'].quantile(0.25)
Q3_carat = df['carat'].quantile(0.75)
IQR_carat = Q3_carat - Q1_carat
lower_bound_carat = Q1_carat - 1.5 * IQR_carat
upper_bound_carat = Q3_carat + 1.5 * IQR_carat
# Find outliers in 'carat'
carat_outliers = df[(df['carat'] < lower_bound_carat) | (df['carat'] > upper_bound_carat)]
carat_outliers
```
