```python
import pandas as pd

import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5, 100)
y1 =x
y2 = x**2

plt.fill_between(x, y1, y2,alpha=0.7, label ='Area between 2 curve' )
plt.legend() # by default label is taken as legend but we can print different legend like .legend('d')
# to show x and y label
plt.xlabel("x-data")
plt.ylabel("y-data")
# to show grid 
plt.grid()
# print the plot
plt.show()
```

```python
# line chart
x = np.linspace(0,5, 100)
y1 =x
y2 = x**2

plt.plot(x, y1, )
plt.legend() # by default label is taken as legend but we can print different legend like .legend('d')
# to show x and y label
plt.xlabel("x-data")
plt.ylabel("y-data")
# to show grid 
plt.grid()
# print the plot
plt.show()
```

```python
x = np.linspace(0,5, 100)
y1 =x
y2 = x**2

plt.fill_between(x,y2,label ='Area Under cover')
plt.legend() # by default label is taken as legend but we can print different legend like .legend('d')
# to show x and y label
plt.xlabel("x-data")
plt.ylabel("y-data")
# to show grid 
plt.grid()
# print the plot
plt.show()
```

```python
x = np.linspace(0,5, 100)
y1 =x
y2 = x**2

plt.fill_between(x, y1, y2,alpha=0.7, label ='Area between 2 curve' )
plt.plot(x,y1,color='green')
plt.plot(x,y2, color='red')
plt.legend() # by default label is taken as legend but we can print different legend like .legend('d')
# to show x and y label
plt.xlabel("x-data")
plt.ylabel("y-data")
# to show grid 
plt.grid()
# print the plot
plt.show()
```

```python
# Pie chart
Labels = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(0,10,len(Labels))

plt.pie(values, labels = Labels, autopct = "%0.2f%%", startangle=120 )
plt.legend() 
# Title
plt.title('Pie Chart Example')
# print the plot
plt.show()
```

```python
# let get the data from csv
df = pd.read_csv("datasets/ADANIPORTS.csv", parse_dates=True)
df
```

```python
# plot 3d plots
plt.figure(figsize=(10,6))

# Take difference of High and Low price 
df['H-L'] = df.High-df.Low
# Calculate moving avaerage on Closing price for 100 days
df['100MA'] = df['Close'].rolling(100).mean()

# plot the 3 axis
ax = plt.axes(projection='3d')
# plot the scatter plot on 3d axis
ax.scatter(df.index, df['H-L'], df['100MA'])

# set the labels for 3 axis
ax.set_xlabel("Index")
ax.set_ylabel("H_L")
ax.set_zlabel("100MA")

plt.show()
```
