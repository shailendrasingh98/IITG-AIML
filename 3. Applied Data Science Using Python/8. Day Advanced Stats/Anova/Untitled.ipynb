```python
import pandas as pd
import numpy as np
import scipy.stats as stats

# Creating sample data
np.random.seed(42)
data = {
    'Team': np.repeat(['India', 'Australia', 'England'], 10),
    'Score': np.concatenate([
        np.random.normal(280, 15, 10),  # India's scores
        np.random.normal(290, 15, 10),  # Australia's scores
        np.random.normal(275, 15, 10)   # England's scores
    ])
}

df = pd.DataFrame(data)

# Perform One-Way ANOVA
f_stat, p_value = stats.f_oneway(df[df['Team'] == 'India']['Score'],
                                 df[df['Team'] == 'Australia']['Score'],
                                 df[df['Team'] == 'England']['Score'])

print(f"F-Statistic: {f_stat:.2f}, P-Value: {p_value:.4f}")
print(f"p-value: {p_value} is less than 0.05 hence it say: Rejection of Null hypothesis")
print(f"Reject of Null hypothesis means there is significance difference between each team ")
```

```python
from scipy.stats import f
k = 3  # Number of groups
N = 30  # Total observations
df_between = k - 1  # 3 - 1 = 2
df_within = N - k   # 30 - 3 = 27
alpha = 0.05  # Significance level
critical_f = f.ppf(1 - alpha, df_between, df_within)
print(f"Critical F-Value: {critical_f:.2f}")
```
