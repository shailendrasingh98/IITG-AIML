**range() in Python**

- it is an in-built function in Python that helps to generate a sequence of numbers.
- It is used to iterate over a sequence of numbers in loops.

*Syntax:*
range(start,stop,step)
- start: accepts an integer value which specifies the starting value of the sequence (optional)
- stop: accepts an integer value which specifies the ending value of the sequence (required) 
 And, the stop value is inclusive only until (stop_value-1)
- step: accepts an integer value which specifies the gap/increment between each number in the sequence to be generated(optional)

```python
range(1,6) # --> this range function is supposed to generate numbers between 0 to 5 with a default step value of 1
```

**for loop**

- i in the for loop is known as the local iterable variable.
- the number of for loop iterations = no of values generated by the range function/present in the collection variable

```python
for i in range(1,6):
    print(i)
```

**How to combine conditions and loops?**

```python
for i in range(1,6):
    if i % 2==0:
        print(i, "is an even number.")
    else:
        print(i, "is an odd number.")
```

**while loop**
- while loop can execute a set of statements only as long as the condition is true.

```python
num = 1

while num < 5:
    print(num)
    num+=1
```

```python
num = 10

while num > 5:
    print(num)
    num-=1
```

**Loop Control statements**

- **break** - abruptly halts the loop execution.
- **continue** - skips the current iteration and it helps to move onto the next iteration.

```python
for i in "Hello":
    if i == "l":
        break
    print(i)

print("End of loop")
```

```python
for i in "Hello":
    if i == "l":
        continue
    print(i)

print("End of loop")
```
