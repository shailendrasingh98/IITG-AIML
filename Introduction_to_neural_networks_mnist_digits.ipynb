In order to normalise our train and test datasets, we divide the pixel values present in every image by 255.

```python
x_train = x_train/255.0
x_test = x_test/255.0
```

```python
print("After data normalisation")
print(x_test[0])
```

One hot encoding the labels

```python
y_train_one_hot = tf.keras.utils.to_categorical(y_train,10)
y_test_one_hot = tf.keras.utils.to_categorical(y_test,10)
```

```python
print("Before OHE")
print(y_train[0])
```

```python
print("After OHE")
print(y_train_one_hot[0])
```

```python
print(x_train.shape)
```

```python
print(x_test.shape)
```

```python
# X_train is 60000 rows of 28x28 values(2d); we reshape it to 60000 x 784(1d)
# X_test is 10000 rows od data 28x28 values(2d); we reshape it to 10000 x 784(1d)
# 28x28 = 784 neurons
x_train_reshaped = x_train.reshape(60000,784)
x_test_reshaped = x_test.reshape(10000,784)
```

Note: We have hence loaded the data and completed all the necessary preprocessing steps.

***Let us begin creating our first Neural Network Model***

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model_mnist_digits = Sequential()
```

```python
model_mnist_digits.add(Dense(10, input_shape=(784,), activation='softmax'))
```

```python
model_mnist_digits.summary()
```

```python
#compile the model
model_mnist_digits.compile(optimizer='SGD', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Training the model

training = model_mnist_digits.fit(x_train_reshaped,y_train_one_hot,epochs=5, validation_split=0.2)
```

```python
#evaluate the model
test_loss, test_acc = model_mnist_digits.evaluate(x_test_reshaped, y_test_one_hot)
print('Test accuracy:', test_acc)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
# list all data in training
print(training.history.keys())
# summarize training for accuracy
plt.plot(training.history['accuracy']) # training accuracy values
plt.plot(training.history['val_accuracy']) #validation accuracy values
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
# summarize traning for loss
plt.plot(training.history['loss']) # training loss values
plt.plot(training.history['val_loss']) #validation loss values
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

```python
#evaluate the model
test_loss, test_acc = model_mnist_digits.evaluate(x_test_reshaped, y_test_one_hot)
print('Test accuracy:', test_acc)
```

Build another model with 1 hidden layer. This model would have 3 layers in total - input, hidden & output layers

```python
model_mnist_digits_hidden = Sequential()
```

```python
model_mnist_digits_hidden.add(Dense(20, activation = "relu", input_shape=(784,))) #input layer with hidden layer
model_mnist_digits_hidden.add(Dense(10, activation = "softmax")) #output layer
```

```python
model_mnist_digits_hidden.summary()
```

```python
model_mnist_digits_hidden.compile(optimizer='SGD', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
training2 = model_mnist_digits_hidden.fit(x_train_reshaped,y_train_one_hot,epochs=5, validation_split=0.2)
```

```python
def plot_graphs(history, string):
  plt.plot(history.history[string])
  plt.plot(history.history['val_'+string])
  plt.xlabel("Epochs")
  plt.ylabel(string)
  plt.legend([string, 'val_'+string])
  plt.title('Model ' + string) # Added a title for clarity

plt.figure(figsize=(12, 5)) # Optional: make the figure wider for two plots
plt.subplot(1, 2, 1) # First subplot for accuracy
plot_graphs(training2, "accuracy")
plt.subplot(1, 2, 2) # Second subplot for loss
plot_graphs(training2, "loss")
plt.tight_layout() # Adjust layout to prevent overlapping plt.show() # Display the entire figure with both subplots
```

# Model 3 : Dropout layer

4 layers in total, 2 hidden layer, 1 input and 1 output, a drop layer in between.:

```python
from tensorflow.keras.layers import Dropout, Input
```

```python
model_3 = Sequential()
model_3.add(Input(shape=(784,)))
model_3.add(Dense(40, name='First_Hidden_Layer', activation = "relu")) # first hidden layer
model_3.add(Dropout(0.2))
model_3.add(Dense(20, name='Second_Hidden_Layer', activation = "relu")) #input layer with hidden layer
model_3.add(Dense(10, name='Output_Layer', activation = "softmax")) #output layer
model_3.summary()
```

```python
model_3.compile(optimizer='SGD', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model_3.fit(x_train_reshaped,y_train_one_hot,epochs=5, validation_split=0.2)
```

```python
#evaluate the model
test_loss, test_acc = model_3.evaluate(x_test_reshaped, y_test_one_hot)
print('Test accuracy:', test_acc)
```

```python
history.history.keys()
```

```python
plt.figure(figsize=(12, 5)) # Optional: make the figure wider for two plots
plt.subplot(1, 2, 1) # First subplot for accuracy
plot_graphs(history, "accuracy")
plt.subplot(1, 2, 2) # Second subplot for loss
plot_graphs(history, "loss")
plt.tight_layout()
```

# Model 6: Batch Normalization
4 layers in total, 2 hidden layer, 1 input and 1 output, a batchnorm layer in between.

```python
from tensorflow.keras.layers import BatchNormalization
```

```python
model_6 = Sequential()
model_6.add(Input(shape=(784,)))
model_6.add(Dense(50, name='First_Hidden_Layer', activation = "relu")) # first hidden layer
model_6.add(BatchNormalization())
model_6.add(Dense(20, name='Second_Hidden_Layer', activation = "relu")) #input layer with hidden layer
model_6.add(Dense(10, name='Output_Layer', activation = "softmax")) #output layer
model_6.summary()
```

```python
model_6.compile(optimizer='SGD', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model_6.fit(x_train_reshaped,y_train_one_hot,epochs=5, validation_split=0.2)
```

```python
test_loss, test_acc = model_6.evaluate(x_test_reshaped, y_test_one_hot)
print('Test accuracy:', test_acc)
```
